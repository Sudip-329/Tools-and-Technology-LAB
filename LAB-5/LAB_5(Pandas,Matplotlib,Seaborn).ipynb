```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('india-crime.csv')
df.head(5)
```

# a) Display total number of Kidnapping and Abduction happened in Maharashtra state.

```python
maharashtra_kidnappings = df[df['STATE/UT'] == 'MAHARASHTRA']['Kidnapping and Abduction'].sum()
print("Total number of Kidnapping and Abduction happened in Maharashtra state:", maharashtra_kidnappings)
```

# b) Display total number of Kidnapping and Abduction happened in Maharashtra state of RAIGAD district.

```python
raigad_kidnappings = df[(df['STATE/UT'] == 'MAHARASHTRA') & (df['DISTRICT'] == 'RAIGAD')]['Kidnapping and Abduction'].sum()
print("Total number of Kidnapping and Abduction happened in Raigad district, Maharashtra state:", raigad_kidnappings)
```

# C) Display the heat map of 7 column(Rape', 'Kidnapping and Abduction',  'Dowry Deaths', 'Assault on women with intent to outrage her modesty',  'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',  'Importation of Girls').

```python
columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
           'Assault on women with intent to outrage her modesty',
           'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
           'Importation of Girls']
subset_df = df[columns]
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(subset_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Crime Categories')
plt.show()
```

```python
columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
           'Assault on women with intent to outrage her modesty',
           'Insult to modesty of Women', 'Cruelty by Husband or his Relatives',
           'Importation of Girls']
subset_df = df[columns]

plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(subset_df, cmap='coolwarm', annot=True, fmt='g')
plt.title('Heatmap of Crime Categories')
plt.show()
```

# d) Display the bar plot Dowry Deaths from 2001 to 2012 in India.

```python
years = range(2001, 2013)
india_years_df = df[df['Year'].isin(years)]

plt.figure(figsize=(9, 5))
plt.bar(india_years_df['Year'], india_years_df['Dowry Deaths'], color='pink')
plt.title('Dowry Deaths in India (2001-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Dowry Deaths')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# E) Display the bar plot Dowry Deaths from 2001 to 2012 in state of west Bengal.

```python
west_bengal_df = df[df['STATE/UT'] == 'WEST BENGAL']
years = range(2001, 2013)
west_bengal_years_df = west_bengal_df[west_bengal_df['Year'].isin(years)]

plt.figure(figsize=(10, 6))
plt.bar(west_bengal_years_df['Year'], west_bengal_years_df['Dowry Deaths'], color='yellow')
plt.title('Dowry Deaths in West Bengal (2001-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Dowry Deaths')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
