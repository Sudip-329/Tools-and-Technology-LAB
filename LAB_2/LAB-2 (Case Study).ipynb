Note: Provide code snippets or pseudo-code for key parts of your implementation, and explain your choices regarding data structures and design decisions.

```python
# Dictionary to store product information
products = {
    '1': {'name': 'Banana', 'price': 40},
    '2': {'name': 'Apple', 'price': 160},
    '3': {'name': 'Mango', 'price': 60},
    '4': {'name': 'Orange', 'price': 100},
    '5': {'name': 'Grape', 'price': 200},
    '6': {'name': 'KIWI', 'price': 300}
}

cart = []

def add_to_cart(product_id):
    if product_id in products:
        cart.append(product_id)
        print("Product added to cart:")
        print(f"{product_id}   {products[product_id]['name']}  {products[product_id]['price']}")
    else:
        print("Invalid product ID. Please enter a valid product ID.")

def remove_from_cart(product_id):
    if product_id in cart:
        cart.remove(product_id)
    else:
        print("Product not found in the cart.")

def view_cart():
    if not cart:
        print("Cart is empty.")
    else:
        print("Items in the cart:")
        for item_id in cart:
            print(products[item_id]['name'])

def calculate_total():
    total = sum(products[item_id]['price'] for item_id in cart)
    return total

def apply_discount(discount_set):
    for item_id in cart:
        if item_id in discount_set:
            products[item_id]['price'] *= 0.9  # 10% discount
    
    discount_total = calculate_total()
    print(f"After Applying discount: {discount_total}")

discount_set = {'1', '2'}

while True:
    print("\n1. Add a product to the cart")
    print("2. Remove a product from the cart")
    print("3. View the contents of the cart")
    print("4. Calculate the total cost of the items in the cart")
    print("5. Apply promotional discounts")
    print("6. Exit")
    
    print("Welcome to the T&T lab Fruit store.")
    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        print("Available products:")
        for key, value in products.items():
            print(f"{key}   {value['name']}   {value['price']}")
        product_id = input("Enter the product ID to add: ")
        add_to_cart(product_id)
    elif choice == '2':
        product_id = input("Enter the product ID to remove: ")
        remove_from_cart(product_id)
    elif choice == '3':
        view_cart()
    elif choice == '4':
        print("Total cost of items in the cart:", calculate_total())
    elif choice == '5':
        apply_discount(discount_set)
    elif choice == '6':
        print("Thanks for shopping with us...")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```
